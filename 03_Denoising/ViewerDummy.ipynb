```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_raw = pd.read_csv('../01_Database_ECGSignal_Rhythm/JS10037.csv')
# ECG I
ecg_signal_raw_I = df_raw.iloc[:, 0].values

df_matlab = pd.read_csv('../01_Database_ECGSignal_Rhythm_Denoised/JS10037.csv')
# ECG I
ecg_signal_matlab_I = df_matlab.iloc[:, 0].values
```

```python
# Time calculation
fs = 500 # Sampling frequency (Hz)
sig_len = 5000 # Signal Length
time = [i / fs for i in range(sig_len)] 
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(24, 8),dpi=150)
plt.plot(ecg_signal_raw_I, label='Raw ECG Signal')
plt.title("Raw ECG Signal (Lead - I)")
plt.xlabel("Time (sec)")
plt.ylabel("Voltage (µV)")
plt.legend()
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(24, 8),dpi=150)
plt.plot(ecg_signal_matlab_I, label='Denoised ECG Signal')
plt.title("Denoised ECG Signal (Lead - I)")
plt.xlabel("Time (sec)")
plt.ylabel("Voltage (µV)")
plt.legend()
plt.show()
```
