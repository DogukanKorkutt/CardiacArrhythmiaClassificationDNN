```python
import pandas as pd
import os

# Folder paths
source_folder = '../02_Diagnosis_PreProcess'
target_folder = '.'

# Source and target files
source_files = [
    'DiagnosticInfo_amplitude.csv',
    'DiagnosticInfo_duration.csv',
    'DiagnosticInfo_morphology.csv',
    'DiagnosticInfo_rhythm.csv',
    'DiagnosticInfo_rhythm_WithoutAddCarCon.csv',
    'DiagnosticInfo_unlabeled.csv'
]

target_files = [
    'ECGSignal_ID_amplitude.csv',
    'ECGSignal_ID_duration.csv',
    'ECGSignal_ID_morphology.csv',
    'ECGSignal_ID_rhythm.csv',
    'ECGSignal_ID_rhythm_WithoutAddCarCon.csv',
    'ECGSignal_ID_unlabeled.csv'
]

# Save the ID of each source file to the target file
for source, target in zip(source_files, target_files):
    # Full path of the source file
    source_path = os.path.join(source_folder, source)
    # Full path of the target file
    target_path = os.path.join(target_folder, target)
    # Read the source file
    df = pd.read_csv(source_path)
    # Get only the ID column
    id_df = df[['ID']]
    # Write the ID column to the target file
    id_df.to_csv(target_path, index=False)
    print(f"{source_path} -> {target_path}: ID variable saved.")
```

```python
import pandas as pd

# List of target files
target_files = [
    'ECGSignal_ID_amplitude.csv',
    'ECGSignal_ID_duration.csv',
    'ECGSignal_ID_morphology.csv',
    'ECGSignal_ID_rhythm.csv',
    'ECGSignal_ID_rhythm_WithoutAddCarCon.csv',
    'ECGSignal_ID_unlabeled.csv'
]

# Initialize an empty list to store the file names and shapes
file_shape_info = []

# Check the shape of each file
for target in target_files:
    file_path = target
    try:
        # Read the file into a DataFrame
        signal_info_df = pd.read_csv(file_path)
        # Get the shape of the dataset
        shape = signal_info_df.shape
        # Append the file name and shape to the list
        file_shape_info.append({'File Name': target, 'Shape': shape})
    except FileNotFoundError:
        # If the file is not found, append a message
        file_shape_info.append({'File Name': target, 'Shape': 'File not found'})

# Create a DataFrame from the list
shape_df = pd.DataFrame(file_shape_info)

# Display the resulting DataFrame
shape_df
```
